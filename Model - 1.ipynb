```python
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
legit_under_sampled = legit.sample(n=492)
```

```python
new_data = pd.concat([legit_under_sampled, fraud], axis=0)
```

```python
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
print(X.shape)
print(Y.shape)
```

```python
X
```

```python
Y
```

```python
k_best = 5
X_selected = SelectKBest(score_func=f_classif, k=k_best).fit_transform(X, Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python


rf_model = RandomForestClassifier(random_state=42)
xgb_model = xgb.XGBClassifier(random_state=42)
avm_model = AdaBoostClassifier(random_state=42)
lr_model = LogisticRegression(random_state=42)





supervised_models = [rf_model, xgb_model, avm_model, lr_model]

model_names_supervised = ["Random Forest", "XGBoost", "AdaBoost", "Logistic Regression"]

results_supervised = []


for model, name in zip(supervised_models, model_names_supervised):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    
    results_supervised.append([name, accuracy, precision, recall, f1, roc_auc])




results_df_supervised = pd.DataFrame(results_supervised, columns=["Model", "Accuracy", "Precision", "Recall", "F1-score", "ROC-AUC"])



print("Supervised Models Results:")
print(results_df_supervised)




```
